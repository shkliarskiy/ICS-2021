### Приклад № 1

Написати програмний код, який по введеному коду фірми поверне на екран 
телефон контактної особи фірми

```python
def find_phone_number():
    ''' функія пошуку телефону по коду фірми
    '''
    firm_code = input('введіть код фірми: ')
    
    with open('firms.txt', encoding='utf-8') as firms_file:
        for firm in firms_file:
            line_list = firm.split(',')
            # перевіити співпадіння кода фірми з заданим
            if line_list[0] == firm_code:
                print(f"Фірма {line_list[1]} - телефон: {line_list[2]}")
                
        print(f'Код {firm_code} не знайдено')

find_phone_number()
```
